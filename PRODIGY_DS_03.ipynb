```python
#Label Encoding Categorical Features
cols = bank.select_dtypes("object").columns
cols
```

```python
le = LabelEncoder()

bank[cols] = bank[cols].apply(le.fit_transform)
bank.head(5)
```

```python
#Correlation Analysis using Heatmap
plt.figure(figsize=(25,10))
sns.heatmap(bank.corr(), cmap='bwr', annot=True)
plt.show()
```

```python
#Splitting input and output
X = bank.drop("y", axis=1)
y = bank.y

# Initialize the StandardScaler
scaler = StandardScaler()

# Standardize the features
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# Describe the scaled features
description = X_scaled.describe()
print(description)
```

```python
#Building the Decision Tree Classifier
#Train-test split
train_X, test_X, train_y, test_y = train_test_split(X_scaled, y, test_size=0.3)

decision_tree = DecisionTreeClassifier()
decision_tree.fit(train_X, train_y)
```

```python
print('Train Score: {}'.format(decision_tree.score(train_X, train_y)))  
print('Test Score: {}'.format(decision_tree.score(test_X, test_y)))    
```

```python
cross_val_score(decision_tree, train_X, train_y, cv=5).mean()
```

```python
ypred = decision_tree.predict(test_X)
print(classification_report(test_y,ypred))
```

```python
#Hyperparameter tunning
#Grid search cv to find best estimaters to improve model performance

param_grid = {
    'max_depth': [3, 5, 7,10, None],
    'criterion' : ['gini', 'entropy'],
    'min_samples_leaf': [3, 5, 7, 9,10,20]
    }

gscv = GridSearchCV(decision_tree, param_grid, cv=5, verbose=1)
gscv.fit(train_X, train_y)
```

```python
gscv.best_params_
```

```python
gscv.best_estimator_
```

```python
cross_val_score(gscv.best_estimator_, train_X, train_y, cv=5).mean() 
```

```python
#Using this best estimators to built our model.
clf = DecisionTreeClassifier(criterion= 'gini', max_depth= 5, min_samples_leaf = 3)
clf.fit(train_X, train_y)
```

```python
print('Train Score: {}'.format(clf.score(train_X, train_y)))
print('Test Score: {}'.format(clf.score(test_X, test_y)))
```

```python
pred_y = clf.predict(test_X)
```

```python
##Confusion Matrix
cm = confusion_matrix(pred_y, test_y)
ConfusionMatrixDisplay(cm, display_labels=clf.classes_).plot()
plt.show()
```

```python
#Classification Report
print(classification_report(pred_y, test_y))
```

```python
#Accuracy Score
accuracy = accuracy_score(test_y,pred_y)
print("Test Accuracy of Decision Tree Classifier : {}".format(accuracy*100))

```

```python
#Cross Validation Score
Cross_val = cross_val_score(clf, test_X,test_y, cv=5).mean()     
print("Cross-Validation Accuracy Scores Decision Tree : ",Cross_val*100)
```

Let's summarize the key insights gained from our analysis:

**Data Pre-processing:**
- Our dataset was well-structured with no missing values. We encountered some duplicate entries but found them to be valid instances of different campaigns targeting the same clients.

**Exploratory Data Analysis (EDA):**
- The majority of clients fell within the 30-40 age range, coinciding with the highest subscription rate for term deposits.
- Occupations such as administrative, blue-collar, and technician roles were prevalent among clients, with a higher likelihood of subscribing to term deposits.
- Married individuals comprised the majority of clients, while those with university degrees or higher education levels showed a higher subscription rate.
- Instances of credit defaults were rare, and clients with housing loans were more likely to subscribe to term deposits compared to those with personal loans.

**Model Building and Evaluation:**
- We developed a Decision Tree classifier that achieved an accuracy rate of approximately 91.46% on both training and testing datasets.
- Cross-validation confirmed the robustness of our model, with an accuracy score of around 90.89%.
- The classification report indicated balanced performance with reasonable precision, recall, and F1-scores for subscribed and non-subscribed clients.

**Insights:**
- Longer call durations were found to be significant predictors of term deposit subscriptions, suggesting the importance of engaging clients effectively during conversations.
- Previous contact attempts also played a role, highlighting the potential effectiveness of repeated contact in convincing clients to subscribe.
- We observed multicollinearity among some input features, which could impact model performance and should be addressed in future iterations. 

Overall, our analysis provides valuable insights for optimizing marketing strategies and enhancing client engagement to increase term deposit subscriptions.
